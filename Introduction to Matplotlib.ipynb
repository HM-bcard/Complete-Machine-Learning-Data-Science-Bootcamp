# A3 Matplotlib example workflow:

```python
# 0. import matplotlib and prepare it for usage inline in jupyter
%matplotlib inline
import matplotlib.pyplot as plt
# 1. Prepare data
x = [1,2,3,4]
y = [33,66,99,132]

# 2. setup plot:

fig,ax = plt.subplots(figsize=(10,10))# width , height

# 3. Plot data

ax.plot(x,y)

# 4. Customizing the plot:
ax.set(title="Di Title",
    xlabel ='x-axis or int 1-4',
    ylabel='y-axis or x axis times 33'
)
# 5, Save and display the figure

fig.savefig("images/sampleplot-section85.png")
```

# A3 __Recreating this plot using the OO method:__

```python
fix,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            cmap = 'RdYlBu');
```

```python
fix,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            cmap = 'RdYlBu',
            ax=ax );
ax.set_xlim([45,100])
```
