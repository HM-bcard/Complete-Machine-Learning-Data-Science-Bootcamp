# A3 Matplotlib example workflow:

```python
# 0. import matplotlib and prepare it for usage inline in jupyter
%matplotlib inline
import matplotlib.pyplot as plt
# 1. Prepare data
x = [1,2,3,4]
y = [33,66,99,132]

# 2. setup plot:

fig,ax = plt.subplots(figsize=(10,10))# width , height

# 3. Plot data

ax.plot(x,y)

# 4. Customizing the plot:
ax.set(title="Di Title",
    xlabel ='x-axis or int 1-4',
    ylabel='y-axis or x axis times 33'
)
# 5, Save and display the figure

fig.savefig("images/sampleplot-section85.png")
```

# A5 Plot from a dict:

```python
nut_butter_price = {"Almond butter":10,
                    "Peanut butter":8,
                    "Cashew butter":12}
fig,ax = plt.subplots()
ax.bar(x=nut_butter_price.keys(),y=nut_butter_price.values()) # doesn't take x and y args
```

```python
fig.savefig("images/butter_price.png")
```

```python
nut_butter_price = {"Almond butter":10,
                    "Peanut butter":8,
                    "Cashew butter":12}
fig,ax = plt.subplots()
ax.bar(nut_butter_price.keys(),nut_butter_price.values())
```
